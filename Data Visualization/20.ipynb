# Seaborn圖樣式
有五種預設seaborn主題：darkgrid，whitegrid，dark，white，和ticks。它們各自適合不同的應用和個人喜好。默認主題是darkgrid

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy.random as rs
```

```python
#設定圖形樣式 - whitegrid
sns.set_style("whitegrid") 

# 利用 NUMPY 去建立資料集
# np.random.RandomState 設定數學式
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)



# 畫圖
# sns.residplot
# plot residuals
sns.residplot(x, y)
```

## 繪製單變數分佈: 快速檢視海出生單變數分佈的最便捷方式是distplot() 函數。默認情況下,這將繪製直方圖並適合內核密度估計值(KDE)。

2. 作業目標(2) 
    1. 使用 distplot()使用簡單的規則來正確猜測預設情況下正確的數位,但嘗試更多或更少的 bin 可能會顯示資料中的其他特徵
    2. 有無kde對圖形分布的影響

```python
# bin: 指的是特徵值, 
# kde: on/off
sns.displot(x, bins=10, kde=True, rug=True)
sns.displot(x, bins=20, kde=False, rug=True)
```
