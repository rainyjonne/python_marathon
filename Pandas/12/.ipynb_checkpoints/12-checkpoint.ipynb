```python
import pandas as pd
import numpy as np
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
df = pd.read_csv("boston.csv")
print(df.boxplot(fontsize=10, rot=45))
print("欄位TAX和B的中位數在300~400之間")
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?

print(df.plot.scatter(x='NOX', y='DIS'))
print("NOX和DIS這兩個欄位呈指數型反比")

```
